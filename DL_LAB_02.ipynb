# This is Deep Learning Lab 2 Exercie
* The task is to create Dense layer class with only forward method, then using that class and random input with (m, 5) shape, making 3 hidden layer each with 16 neurons and final output layer with 5 neuros.

```python
import torch
```

```python
class DenseLayer:
  def __init__(self, n_features, n_neurons):
    self.weights = 0.01 * torch.rand((n_features, n_neurons))
    self.bias = torch.rand(1, n_neurons)

  def forward(self, inputs):
    self.output = torch.matmul(inputs, self.weights) + self.bias
```

```python
inputs = torch.rand((200, 5))
```

```python
layer1 = DenseLayer(5, 16)
layer2 = DenseLayer(16, 16)
layer3 = DenseLayer(16, 16)
outputClass = DenseLayer(16, 5)
```

```python
layer1.forward(inputs)
layer2.forward(layer1.output)
layer3.forward(layer2.output)
outputClass.forward(layer3.output)
```
